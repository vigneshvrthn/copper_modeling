```python
#importing the required packages 
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
mydb=sqlite3.connect("coppermodel.db")
mycursor=mydb.cursor()
```

```python
#retrive the datas from the sql
df=pd.read_sql_query("select * from copper",mydb)
df
```

```python
# Function to chech the skewness by using the thershold
def interpret_skewness(value):
    if value < -1:
        return "Highly left-skewed"
    elif -1 <= value < -0.5:
        return "Moderately left-skewed"
    elif -0.5 <= value <= 0.5:
        return "Approximately symmetric"
    elif 0.5 < value <= 1:
        return "Moderately right-skewed"
    else:
        return "Highly right-skewed"

```

```python
# in the data the except the catogorical we taken the continious coloumn to ckeck the skewness
for i in ["quantity_tons","thickness","width","selling_price"]:
    sns.distplot(df[i], kde=True, bins=30)  # KDE stands for Kernel Density Estimation
    skewness_quantity_tons = df[i].skew()
    interpretation = interpret_skewness(skewness_quantity_tons)

    

    plt.title("Distribution of " + i+interpretation)
    plt.xlabel(i)
    plt.ylabel("Density")
    plt.show()
    
```
